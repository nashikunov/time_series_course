Я взяла акции LVMH Moët Hennessy — Louis Vuitton https://finance.yahoo.com/quote/MC.PA?p=MC.PA

История цены акции с 2000 года

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
plt.style.use('bmh')
```

```python
data = pd.read_csv('MC.PA.csv', parse_dates=['Date'])
```

```python
data.dropna(inplace=True)
```

```python
data.head()
```

```python
plt.figure(figsize=(16, 8))
plt.title('Акции LVMH')
plt.xlabel('Время')
plt.ylabel('$')
plt.plot(data.Date, data.Close)
plt.show()
```

```python
fft_coef = np.fft.rfft(data.Close)
ps_coef = np.abs(fft_coef)**2
```

```python
log_x = np.log(np.arange(1, len(ps_coef)+1))
log_y = np.log(ps_coef).reshape(-1, 1)
```

```python
lr = LinearRegression()
lr.fit(log_x.reshape(-1, 1), log_y.reshape(-1, 1))
```

```python
lr.coef_
```

# Угол наклона -1.89

```python
plt.figure(figsize=(16, 8))
plt.plot(log_x, log_y, label='Cпектральная плотность мощности в масштабе log')
plt.plot(log_x, lr.predict(log_x.reshape(-1, 1)), label=f'Апроксимация плотности мощности', linewidth=3)
plt.legend()
plt.show()
```

Акции Apple https://finance.yahoo.com/quote/AAPL/history?p=AAPL

Глубина данных 1980-2020

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('AAPL.csv')
```

```python
df.dropna(inplace=True)
```

```python
df.Date = pd.to_datetime(df.Date)
```

```python
df.head()
```

```python
plt.figure(figsize=(15, 8))
plt.title('Акции Apple 1980-2020')
plt.plot(df.Date, df.Close)
plt.show()
```

```python
fft_coef = np.fft.rfft(df.Close)
ps_coef = np.abs(fft_coef)**2
```

```python
len(ps_coef)
```

```python
log_x = np.log(np.arange(1, 4952))
log_y = np.log(ps_coef)
```

```python
lr = LinearRegression()
lr.fit(log_x.reshape(-1, 1), log_y.reshape(-1, 1))
```

```python
print(lr.coef_[0][0])
```

## Угол наклона -1.853

```python
prediction = lr.predict(log_x.reshape(-1, 1))
```

```python
plt.figure(figsize=(15, 8))
plt.plot(log_x, log_y)
plt.plot(log_x, prediction, linewidth=2)
plt.show()
```

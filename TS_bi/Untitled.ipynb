```python
k = 1000
```

```python
N = 100000
```

```python
random_array = np.random.randn(k, N)
```

```python
y = np.random.randn(k, N).cumsum(axis=1)
```

```python
for el in y:
    plt.plot(el, linewidth=0.2)
plt.plot(y.mean(axis=0), linewidth=4)
plt.plot(y.std(axis=0), linewidth=4)
plt.show()
```
